```python
# read csv file and store as list of data points
import csv
import sys
import random

f = open('sobar-72.csv')
csv_f = csv.reader(f)
colh = next(csv_f)
dat = []

for r in csv_f:
    dat += [r]

f.close()
```

```python
# normalizing dataset
# a small note : we dont normalize as (x-u)/sigma instead we just do (x)/sigma as while calculating distance between two
# points we take their difference which automatically cancels u because (x1-u)/sigma - (x2-u)/sigma = (x1-x2)/sigma

sig = [] # standard deviation

for i in range(19): # 19 variables/features
    mu = 0 # mean
    mu2 = 0 # mean of squares
    n = 0
    for it in dat:
        it[i] = int(it[i])
        mu += it[i]
        mu2 += it[i]*it[i]
        n = n + 1
    mu /= n
    mu2 /= n
    sig += [mu2 - mu**2]

    
for i in range(19):
    for it in dat:
        it[i] /= sig[i]
```

```python
# shuffling and getting test data and training data

random.shuffle(dat) # initial shuffling of data  
tr_dat = dat[0:57] # training data
test_c = dat[57:72] # test cases (last 15)
cset = [] # condensed training data
```

```python
# applying condensation algorithm
  

cset = [tr_dat[0]]
random.shuffle(tr_dat)   
    
# condensing training data to condensed set
while True:
    flag = 0 # check whether cset changed in last iteration
    
    for it in tr_dat: # iterate training data
        cld = sys.maxsize # closest neighbour
        clp = None # reference to data point
        
        for cit in cset: # iterate condensed data currently made
            dist = 0
            for i in range(19): # evaluate distance
                dist += (cit[i] - it[i]) ** 2
            if (dist < cld):
                cld = dist
                clp = cit
        
        if (clp[19] != it[19]):
            cset += [it]
            flag = 1
            break
            
    if (flag):
        continue
    else:
        break
```

```python
# evaluating accuracy of CNN
cnt  = 0
for it in test_c:
    
    parity = 0
    
    #first nearest
    near1 = sys.maxsize
    near1p = None
    #second nearest
    near2 = sys.maxsize
    near2p = None
    #third nearest
    near3 = sys.maxsize
    near3p = None
    #fourth nearest
    near4 = sys.maxsize
    near4p = None
    #fifth nearest
    near5 = sys.maxsize
    near5p = None
    
    for cit in cset:
        dist = 0
        for i in range(19):
            dist += (cit[i] - it[i]) ** 2
        if(dist<near1):           
            near5 = near4
            near5p = near4p           
            near4 = near3
            near4p = near3p           
            near3 = near2
            near3p = near2p           
            near2 = near1
            near2p = near1p            
            near1 = dist
            near1p = cit
        elif(dist<near2):           
            near5 = near4
            near5p = near4p           
            near4 = near3
            near4p = near3p           
            near3 = near2
            near3p = near2p           
            near2 = dist
            near2p = cit            
        elif(dist<near3):           
            near5 = near4
            near5p = near4p           
            near4 = near3
            near4p = near3p           
            near3 = dist
            near3p = cit           
        elif(dist<near4):           
            near5 = near4
            near5p = near4p           
            near4 = dist
            near4p = cit
        elif(dist<near4):           
            near5 = dist
            near5p = cit
        else:
            continue
     
    parity = int(near1p[19]) + int(near2p[19]) + int(near3p[19]) + int(near4p[19]) + int(near5p[19])
    if(parity>=3 and it[19]!='1'):
        cnt += 1
    elif(parity<3 and it[19]!='0'):
        cnt += 1

acc = (100 - cnt * 100 / 15)
print(acc)
f = open("output.txt", "w")
f.write("Accuracy of CNN: "+str(acc)+'\n')
```
